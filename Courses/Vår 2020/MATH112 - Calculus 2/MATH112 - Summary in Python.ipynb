## **Solving integrals:**

```python
# with sympy:

# Define variables:
x, y, z = sy.symbols('x y z')

# Define function:
f = x * sy.sqrt(1+(4*x**2))

# Calculate:
sy.integrate(f, (z, 0, 2), (x, 0, 3))
```

```python
# With numpy and scipy:

# Define function:
f = lambda x: x * sy.sqrt(1+(4*x**2))

# Calculate:
integrate.nquad(f, [[0, 2],[0, 3]])

# nquad(f, [[0, 2],[0, 3])
# See in fraction:
# str(Fraction(I(2)[0]))
```

## Plotting 2d or 3d:

```python
# Range of t:
lower_t = 0
upper_t = np.pi
n = 100
t = np.linspace(lower_t, upper_t, num=n)

# Parametrized equations:
x = np.cos(2*t)
y = np.sin(2*t)

# Plots the grapgh:
plt.plot(x, y)

# set up the range for the axis
# plt.ylim(0, 100) 
# plt.xlim(0, 0)
```

```python
def z(x, y):
    a = np.zeros((40,40))
    return a+np.sqrt(y-x-2) if y>x+2 else a

x = np.arange(1, 4, 0.1)
y = np.arange(1, 4, 0.1)
X, Y = np.meshgrid(x, y)
Z = z(X,Y)

print(Z)
```

```python
# Setting the axes at the center:
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# 2-dimension line plot:
ax.plot(x,y)
plt.show()
```

```python
# 2-dimension line plot:
ax = plt.axes(projection='3d')
ax.plot(z, x, y, "red")
plt.show()
```

## Three-dimensional Points and Lines

```python
# Line plot where xs, ys, zs are the points and zdir direction:
zline = np.linspace(0, 15, 1000)
xline = np.cos(zline)
yline = np.sin(zline)

print(zline.shape, xline.shape, yline.shape)
```

```python
# line plot:
ax = plt.axes(projection='3d')
ax.plot(zline, xline, yline, "red")

# points (scatter) plot:
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter(zdata, xdata, ydata, c=zdata);

plt.show()
```

## Three-dimensional Contour Plots

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

# coordinate vectors:
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

# np.meshgrid: return coordinate matrices from coordinate vectors:
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

print(x.shape, y.shape, X.shape, Y.shape, Z.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# set the elevation (60) and azimuthal (35) angles;
# returns rotated fig with 35 degrees counter-clockwise about the z-axis:
ax.view_init(60, 35)
fig
```

## Wireframes and Surface Plots:

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('wireframe')
plt.show()
```

```python
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface')
plt.show()
```

```python
# Creates a partial polar grid:
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)

X = r * np.sin(theta)
Y = r * np.cos(theta)
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none');
plt.show()
```

## Surface Triangulations

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)

print(theta.shape, r.shape, x.shape, y.shape, z.shape)
```

```python
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, c=z, cmap='viridis', linewidth=0.5)
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z,
                cmap='viridis', edgecolor='none')
plt.show()
```
